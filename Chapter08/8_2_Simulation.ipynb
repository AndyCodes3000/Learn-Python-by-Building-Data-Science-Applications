## Visualisation

```python
datas = {"Heaven Islands":stats, 
        'Harsh Islands':h_stats}

colors = {
    'Heaven Islands': 'blue',
    'Harsh Islands': 'red'
}
```

```python
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], color=v, lw=4, label=k) for k, v in colors.items()]
```

```python
fig, axes = plt.subplots(4, figsize=(10,10), sharex=True)


for i, title in enumerate(('Population', 'Average age', 'Average Survival Skill', '% with SSK > 75')):
    axes[i].set_ylabel(title)
    axes[i].set_xlim(0, 15)

for k, v in datas.items():
    
    for s in v: # for each island
        years = list(s.keys())
        
        axes[0].plot(years, [v['pop'] for v in s.values()], c=colors[k], label=k, alpha=.007)
        axes[1].plot(years, [v.get('mean_age', None) for v in s.values()], c=colors[k], label=k, alpha=.003)
        axes[2].plot(years, [v.get('mean_skill', None) for v in s.values()], c=colors[k], label=k, alpha=.007)
        axes[3].plot(years, [v.get('75_skill', None) for v in s.values()], c=colors[k], label=k, alpha=.007)
        
plt.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 4.65));
```
