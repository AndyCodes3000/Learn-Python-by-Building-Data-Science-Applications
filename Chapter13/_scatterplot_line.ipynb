```python
import pandas as pd
import numpy as np

%matplotlib inline
import pylab as plt
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('./data/EF_battles_corrected.csv', parse_dates=['start', 'end'])
```

```python
from numpy.polynomial.polynomial import polyfit
mask = data[['allies_tanks', 'allies killed']].isnull().any(1)

b, c = polyfit(data.loc[~mask, 'allies_tanks'], data.loc[~mask, 'allies killed'], 1)
X = data.loc[~mask, 'allies_tanks'].sort_values()
```

```python
ax = data.plot(kind='scatter', x='allies_tanks', y='allies killed', figsize=(10,10), c='k')
ax.plot(X, b + c * X, '-', c='red', linewidth=2, alpha=.3)
```
